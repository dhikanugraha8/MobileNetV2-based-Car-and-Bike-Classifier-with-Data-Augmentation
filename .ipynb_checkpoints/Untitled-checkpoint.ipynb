## 2. Importing Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

## 3. Data Preparation

### 3.1 Load Dataset

```python
path_dir = './data/dataset'

# Split data training and validation
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    path_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(224, 224),  # Sesuai ukuran input MobileNet
    batch_size=32
)

validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    path_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(224, 224),  # Sesuai ukuran input MobileNet
    batch_size=32
)
```

```python
# Display samples from training and validation datasets
def display_sample(dataset, num_samples=5):
    for images, labels in dataset.take(num_samples):
        for i in range(num_samples):
            plt.subplot(1, num_samples, i + 1)
            plt.imshow(images[i].numpy().astype("uint8"))
            plt.title(f"Class: {labels[i].numpy()}")
            plt.axis("off")
    plt.show()

print("Training Dataset:")
display_sample(train_dataset)

print("Validation Dataset:")
display_sample(validation_dataset)

class_names = train_dataset.class_names
print(f"Class Name : {class_names}")
```

### 3.2 Data Augmentation

```python
# Agumentation image with flip and rotation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip('horizontal'),
    tf.keras.layers.RandomRotation(0.2)])
train_dataset = train_dataset.map(lambda x, y: (data_augmentation(x, training=True), y))
```

## 4. Model Development

### 4.1 Load Pre-trained Model

```python
# Load the pre-trained MobileNetV2 model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

### 4.2 Fine-tuning the Model

```python
# Freeze the base model layers except for the last 10 for fine tuning
for layer in base_model.layers[:-10]:
    layer.trainable = False
```

### 4.3 Custom Layers

```python
# Add custom classification layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x) 
predictions = Dense(2, activation='softmax')(x) 

# Create the final model
model = tf.keras.Model(inputs=base_model.input, outputs=predictions)
```

### 4.4 Model Compilation

```python
# Compile the model with adam optimizer
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Summary layer model mobilenetv2 with fine tuning
model.summary()
```

## 5. Model Training

```python
# Train the model
history = model.fit(
    train_dataset,
    epochs=30,
    validation_data=validation_dataset
)
```

## 6. Model Evaluation

```python
# Model evaluate
test_loss, test_acc = model.evaluate(validation_dataset)
print('Test loss:', test_loss)
print('Test accuracy:', test_acc)
```

```python
# Display train and validation accuracy each epoch
fig1 = plt.gcf()
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.axis(ymin=0.4,ymax=1)
plt.grid()
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'validation'])
plt.show()
```

## 7. Predictions

```python
# Load image for predict
img_paths = ['data/predict/bike.jpg', 'data/predict/car.jpg', 'data/predict/car-bike.jpg'] 
images = []

for img_path in img_paths:
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    images.append(x)

# Stack all images into a single array for prediction
images = np.vstack(images)

# Visualize the three images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
for i, img_path in enumerate(img_paths):
    img = image.load_img(img_path, target_size=(224, 224))
    ax[i].imshow(img)
    ax[i].axis('off')
    ax[i].set_title(f"Image {i+1} for Prediction")
plt.show()

# Predict the classes of the three images
classes = model.predict(images)
print("Prediction results (probabilities for each class):\n", classes)

# Set the class names in the correct order
class_names = ['Bike', 'Car']

# Display the prediction results and probabilities for each image
for i in range(len(img_paths)):
    predicted_class = class_names[np.argmax(classes[i])]
    confidence = np.max(classes[i])
    
    print(f"\nPrediction for Image {i+1}: {predicted_class}")
    print(f"Confidence Level: {confidence * 100:.2f}%")
    
    print("Class probabilities:")
    for j, class_name in enumerate(class_names):
        print(f"{class_name}: {classes[i][j] * 100:.2f}%")
```
